# Let's make a real time perfect model

```python
#load dataset
dataf = pd.read_csv("M:\Machine Learning Start with CIT\GitHub Projects Machine Learning Algorithm\Data files\Data.csv")
```

```python
#split feature and Target
x = dataf[['height']]
y = dataf[['weight']]
```

```python
#Train test split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(f"X_train Shape: {xtrain.shape}")
print(f"X_tesr Shape: {xtest.shape}")
print(f"y_train Shape: {ytrain.shape}")
print(f"X_test Shape: {xtest.shape}")
```

```python
#model implementation
linearR = LinearRegression()
```

```python
main_model = linearR.fit(xtrain, ytrain)
```

```python
#check slope and intercept
print(f"Slope (m): {main_model.coef_}")
print(f"Intercept (c): {main_model.intercept_}")
```

```python
#check the score
print(f"Training score: {main_model.score(xtrain, ytrain)}")
print(f"Testing score: {main_model.score(xtest, ytest)}")
```

```python
#predict generate
predict_y_m = main_model.predict(xtest)
```

```python
#mae check 
mae1 = mean_absolute_error(ytest, predict_y_m)
mae1
```

```python
#mse check 
mse1 = mean_squared_error(ytest, predict_y_m)
mse1
```
