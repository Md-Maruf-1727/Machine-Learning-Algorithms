```python
#Scatter plot
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="experience_years", y="salary", color="green", size=100)
plt.xlabel("Experience Year")
plt.ylabel("Salary")
plt.title("Relation Between Expeience year and Salary")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="age", y="salary", color="green", size=100)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Relation Between Age and Salary")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="education_level", y="salary", color="green", size=100)
plt.xlabel("Educationa Level")
plt.ylabel("Salary")
plt.title("Relation Between Educationa Level and Salary")
plt.show()
```
