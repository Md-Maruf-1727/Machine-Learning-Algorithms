```python
port_st = PorterStemmer()
sentnce = "Natural Language Processing (NLP) is a field of artificial intelligence that helps computers understand human language."
sentnce_token = word_tokenize(sentnce)
stem_word = [port_st.stem(word2) for word2 in sentnce_token]
insen = " ".join(stem_word)
insen
```

```python
new_sentence = "Natural Language Processing is fun. It helps computers understand human language. People use NLP in chatbots. It is also useful for sentiment analysis. Many companies use NLP to improve their products. Learning NLP opens new opportunities."
new_sentence_port_st = PorterStemmer()
new_sentence_token = word_tokenize(new_sentence)
for new_sentence_stem in new_sentence_token:
    new_sentence_token_stemmer = new_sentence_port_st.stem(new_sentence_stem)
    print(f'{new_sentence_stem} --> {new_sentence_token_stemmer}')
```

```python
wnl_sen = "Natural Language Processing is fun. It helps computers understand human language. People use NLP in chatbots. It is also useful for sentiment analysis. Many companies use NLP to improve their products. Learning NLP opens new opportunities."
wnl = WordNetLemmatizer()
wnl_sen_token = word_tokenize(wnl_sen)
lem_word = [wnl.lemmatize(wnl_sen_token_lem) for wnl_sen_token_lem in wnl_sen_token]
join_lem_word = " ".join(lem_word)
join_lem_word
```

```python
wnl_sen_new = "Natural Language Processing is fun. It helps computers understand human language. People use NLP in chatbots. It is also useful for sentiment analysis. Many companies use NLP to improve their products. Learning NLP opens new opportunities."
wnl_sen_new_token = word_tokenize(wnl_sen_new)
for wnl_sen_new_lem in wnl_sen_new_token:
    wnl_sen_new_token_lematize = wnl.lemmatize(wnl_sen_new_lem)
    print(f'{wnl_sen_new_lem} --> {wnl_sen_new_token_lematize}')
```
