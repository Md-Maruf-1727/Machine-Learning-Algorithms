```python
sent = "Natural Language Processing (NLP) is a field of artificial intelligence that helps computers understand human language."
```

```python
word = word_tokenize(sent)
```

```python
for ind, words in enumerate(word):
    print(f'{ind}: {words}')
```

```python
para = "Natural Language Processing is fun. It helps computers understand human language. People use NLP in chatbots. It is also useful for sentiment analysis. Many companies use NLP to improve their products. Learning NLP opens new opportunities."
```

```python
sent_tok = sent_tokenize(para)
sent_tok
```

```python
for indx, senten in enumerate(sent_tok):
    print(f'{indx}: {senten}')
```
